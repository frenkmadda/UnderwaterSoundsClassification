```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Plot delle metriche di training sul classificatore binario effettuato con GoogleNet

```python
# Carica il file CSV
file_path = 'results/training_metrics.csv'
df = pd.read_csv(file_path)

# Definisci le metriche da plottare
metrics = df.columns[1:]

# Plot di tutte le metriche
plt.figure(figsize=(16, 12))

for i, metric in enumerate(metrics, 1):
    plt.subplot(4, 3, i)
    plt.plot(df['epoch'], df[metric], label=metric)
    plt.xlabel('Epoch')
    plt.ylabel(metric)
    plt.title(metric)
    plt.legend()

plt.tight_layout()
plt.show()
```

```python

# Carica il file CSV
file_path = 'results/training_metrics.csv'
df = pd.read_csv(file_path)

# Plot Precision vs Recall per il training e la validazione separatamente
plt.figure(figsize=(12, 6))

# Training Precision vs Recall
plt.subplot(1, 2, 1)
plt.plot(df['train_recall'], df['train_precision'], marker='o', color='blue', label='Train')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Training Precision vs Recall')
plt.legend()
plt.grid(True)

# Validation Precision vs Recall
plt.subplot(1, 2, 2)
plt.plot(df['val_recall'], df['val_precision'], marker='x', color='red', label='Validation')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Validation Precision vs Recall')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
