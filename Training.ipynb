```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_metrics(file_path):
    # Carica il file CSV
    df = pd.read_csv(file_path)

    # Per ogni metrica nel DataFrame (esclusa la colonna 'epoch')
    for metric in df.columns[1:]:
        if 'train' in metric:
            # Crea la figura
            plt.figure(figsize=(10, 6))

            # Plot della metrica di training
            plt.plot(df['epoch'], df[metric], label=metric, color='blue')

            # Plot della corrispondente metrica di validation
            val_metric = metric.replace('train', 'val')
            if val_metric in df.columns:
                plt.plot(df['epoch'], df[val_metric], label=val_metric, color='red')

            # Imposta le etichette degli assi e il titolo
            plt.xlabel('Epoch')
            plt.ylabel(metric.split('_')[1])
            plt.title(f'Train vs Validation {metric.split("_")[1]}')

            # Mostra la legenda
            plt.legend()

            # Mostra il grafico
            plt.show()
```

```python
def plot_precision_recall(file_path):
    # Carica il file CSV
    df = pd.read_csv(file_path)

    # Plot Precision vs Recall per il training
    plt.figure(figsize=(6, 6))
    plt.plot(df['train_recall'], df['train_precision'], marker='o', color='blue', label='Train')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Training Precision vs Recall')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plot Precision vs Recall per la validation
    plt.figure(figsize=(6, 6))
    plt.plot(df['val_recall'], df['val_precision'], marker='x', color='red', label='Validation')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Validation Precision vs Recall')
    plt.legend()
    plt.grid(True)
    plt.show()
```

# Plot delle metriche di training sul classificatore binario con average=binary effettuato con GoogleNet

```python
plot_metrics('results/training_metrics.csv')
```

```python
plot_precision_recall('results/training_metrics.csv')
```
